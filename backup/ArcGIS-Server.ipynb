```python
import arcpy
import os

# Define the path to the ArcGIS Pro project file
print('begin')
models_base_directory = '/data/MyDataBase/SWATGenXAppData/SWATplus_by_VPUID/huc12/'
project_base_directory = "/data/MyDataBase/SWATGenXAppData/ArcGIS_Server/Models_presentation/Models_presentation/"

# Open an existing project
project_name = 'Models'
# Base directory and project directory
os.makedirs(project_base_directory, exist_ok=True)

aprx = arcpy.mp.ArcGISProject(f"{project_base_directory}/{project_name}.aprx")

# List of names based on the sub-folders under the base directory
NAMES = os.listdir('/data/MyDataBase/SWATGenXAppData/SWATplus_by_VPUID/huc12/')
NAMES.remove('log.txt')

# Get references to the group layers
new_map = aprx.listMaps()[0]

subbasins_group = None
streams_group = None
for group_layer in new_map.listLayers():
    if group_layer.isGroupLayer:
        if group_layer.name == 'subbaisins_group':
            subbasins_group = group_layer
        if group_layer.name == 'streams_group':
            streams_group = group_layer

# Loop over each NAME to move existing feature layers to group layers
for NAME in NAMES[0:12]:
    print(NAME)

    # Find the existing subbasin and stream layers in the map
    subbasin_layer = None
    stream_layer = None
    for layer in new_map.listLayers():
        if layer.isFeatureLayer:
            if os.path.basename(layer.dataSource) == 'SWAT_plus_subbasins_modified.shp':
                subbasin_layer = layer
            elif os.path.basename(layer.dataSource) == 'rivs1.shp':
                stream_layer = layer

    # Move the layers to the respective group layers
    if subbasins_group and subbasin_layer:
        subbasins_group.addLayer(subbasin_layer)
    if streams_group and stream_layer:
        streams_group.addLayer(stream_layer)

    aprx.save()  # Save the project after moving the layers

# Clean-up resources
del aprx
```
