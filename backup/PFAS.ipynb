```python
import pandas as pd
import numpy as np
from shapely import Point 
import geopandas as gpd
import matplotlib.pyplot as plt
pfas_path = r"D:\MyDataBase\PFAS_HuronRiver\2023-12-18 - Data for DSA EGLE-MSU for MDNR 200000000672.xlsx"
CWS_Well_Locations = pd.read_excel(pfas_path, sheet_name = 'CWS Well Locations')
NTNC_Well_Locations = pd.read_excel(pfas_path, sheet_name = 'NTNC Well Locations')
Wellogic_Wells = pd.read_excel(pfas_path, sheet_name = 'Wellogic Wells')
```

```python
PFAS_Sample_Locations = pd.read_excel(pfas_path, sheet_name = 'PFAS Sample Locations')
PFAS_Sample_Locations['geometry'] = [Point(x,y) for (x,y) in zip(PFAS_Sample_Locations.Longitude,PFAS_Sample_Locations.Latitude) ]
PFAS_Sample_Locations = gpd.GeoDataFrame(PFAS_Sample_Locations, geometry='geometry', crs='EPSG:4326')
PFAS_Sample_Locations = PFAS_Sample_Locations.replace(np.nan, 0)
#PFAS_Sample_Locations.plot(column='PFOSResult')

plt.bar(PFAS_Sample_Locations.index, PFAS_Sample_Locations['PFOSResult'])
plt.xlabel('Sample Well Index')
plt.ylabel('PFOS Result')
plt.title('PFOS Results for Each Well')
plt.show()
```
