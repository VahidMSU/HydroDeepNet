```python

############################################################## clean up previous model evaluations ###########################################
import os
import pandas as pd
import importlib
import shutil 

def delete_previous_runs(directory_path):
    if not os.path.exists(directory_path):
        return
    folders_in_directory = os.listdir(directory_path)
    for folder_name in folders_in_directory:
        try:
            if folder_name not in ['Default']:
                folder_path = os.path.join(directory_path, folder_name)
                if os.path.isdir(folder_path):
                    try:
                        shutil.rmtree(folder_path)
                    except Exception as e:
                        print(f'Not deleted: {folder_path}, due to error: {e}')

        except Exception as e:
            print(f'Not deleted: {folder_path}, due to error: {e}')

BASE_PATH = "/data/MyDataBase/SWATGenXAppData/"
NAMES=os.listdir(os.path.join(BASE_PATH, "SWAT_input/huc12/"))
NAMES.remove('log.txt')
MODEL_NAMES = ["SWAT_gwflow_MODEL","SWAT_MODEL","SWAT_gwflow_snow_MODEL"]
LEVEL ='huc12'
for NAME in NAMES:
    for MODEL_NAME in MODEL_NAMES:
        model_base= f'SWAT_input/{LEVEL}/{NAME}/'
        scenarios_path           = os.path.join(BASE_PATH, model_base, f'{MODEL_NAME}/Scenarios')
        delete_previous_runs(scenarios_path)
        
```

```python
#correcting SWAT+ models for wrong elevation values in hru.con file
# 
import os
import pandas as pd
import numpy as np
BASE_PATH = "/data/MyDataBase/SWATGenXAppData/"
LEVELS = ['huc8','huc12']


MODEL_NAMES = ['SWAT_MODEL', 'SWAT_gwflow_MODEL', 'SWAT_gwflow_snow_MODEL']

for LEVEL in LEVELS:
    NAMES = os.listdir(os.path.join(BASE_PATH,'SWATplus_by_VPUID',LEVEL)) 
    if 'log.txt' in NAMES:
        NAMES.remove('log.txt')
    
    for NAME in NAMES:
        for MODEL_NAME in MODEL_NAMES:
          #  if MODEL_NAME == 'SWAT_gwflow_MODEL' and NAME == '41000130101':
            print('yes')
            SOURCE = os.path.join(BASE_PATH, f"SWATplus_by_VPUID/{LEVEL}/{NAME}/{MODEL_NAME}/Scenarios/Default/TxtInOut/hru.con")
            if not os.path.exists(SOURCE):
                print(f'{LEVEL} {NAME} {MODEL_NAME} does not exists')
                continue
            SOURCE_corrected = os.path.join(BASE_PATH, f"SWATplus_by_VPUID/{LEVEL}/{NAME}/{MODEL_NAME}/Scenarios/Default/TxtInOut/hru.con")
            hru_con = pd.read_csv(SOURCE, skiprows=1, sep='\s+')

            median_elev = hru_con[(hru_con.elev>0) & (hru_con.elev<1e3)].elev.median()

            if len(hru_con[hru_con.elev<0])>0 or len(hru_con[hru_con.elev>1e3])>0:
                print(f'error {NAME} median elev: {median_elev}')

                hru_con['elev'] = np.where(hru_con['elev']<0, median_elev,hru_con['elev'] )
                hru_con['elev'] = np.where(hru_con['elev']>1e3, median_elev,hru_con['elev'] )

                with open(SOURCE_corrected, 'w') as file:
                    file.write('corrected elev Vahid Rafiei\n ')
                #hru_con.to_csv(SOURCE_corrected, mode='a', index=False, sep=' ')
                print(SOURCE_corrected)
```
